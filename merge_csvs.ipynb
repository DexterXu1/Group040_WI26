```python
import pandas as pd
import os # Use this to grab csv files in 00-raw

raw_folder = "data/00-raw"
skipped_file = "capes_data.csv" # csv file we are not using

df_list = [] # list of all the csv files read as a dataframe

for files in os.listdir(raw_folder): # os.listdir returns name of all the files in that folder
    if files != skipped_file and files.endswith(".csv"): # since we are not merging capes_data.csv file 
        filename = os.path.join(raw_folder, files) # concatenate folder path name and file name into a path
        
        df = pd.read_csv(filename) # create a dataframe with the files

        csv_file_name = files.replace(".csv", "") # name of the csv files without the .csv at the end
        last_underscore = csv_file_name.rfind("_") # get the last underscore in order to get class name and quarter by separating them
        class_name = csv_file_name[:last_underscore] 
        quarter = csv_file_name[last_underscore + 1:] 

        df["class"] = class_name # create class column
        df["quarter"] = quarter # create quarter column

        df_list.append(df) # adds df to a list of all the dataframes

merged_df = pd.concat(df_list) # merges into one singular dataframe
merged_df.to_csv("data/01-interim/merged_classes.csv") # saves merged dataframe into 01-interim folder
```
